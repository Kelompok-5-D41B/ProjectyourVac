```python
from selenium import webdriver
import time
```

```python
import json
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.traveloka.com/id-id/activities/search?st=GEO&eid=103859&funnel_source=Merchandising&funnel_id=21420&internal_source=true"
```

```python
driver.get(url)
print(driver.current_url)
driver.implicitly_wait(30)
```

```python
for i in range(1, 5):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(15)
```

```python
for results in driver.find_elements_by_xpath("//div[@class='css-1dbjc4n r-13qz1uu']"):
     print(results.text)
```

```python
results = driver.find_elements_by_xpath("//div[@class='css-1dbjc4n r-13qz1uu']")
len(results)
```

```python
Traveling = []
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
for product in driver.find_elements_by_xpath("//div[@class='css-1dbjc4n r-13qz1uu']"):
    if product.text != '':
        images = []
        for img in product.find_elements_by_tag_name('img'):
            images.append({'alt': img.get_attribute('alt'), 'src': img.get_attribute('src')})
        Traveling.append({
            'brand_img': images[0]['src'],
        })
```

```python
len(Traveling)
```

```python
Traveling
```

```python
driver.quit()
```

```python
with open('TravelokaImage.json', 'w') as file:
    json.dump(Traveling, file, indent=1)
```
