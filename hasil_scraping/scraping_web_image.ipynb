```python
import time
import json
from pprint import pprint
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
```

```python
chrome_path = "\Program Files\chromedriver_win32\chromedriver"
```

```python
page_url = "https://www.merdeka.com/gaya/tempat-wisata-di-bandung-kln.html"
```

```python
driver = webdriver.Chrome(executable_path=chrome_path)
driver.get(page_url)
time.sleep(2)
```

```python
for i in range(1, 5):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(15)
```

```python
for results in driver.find_elements_by_xpath("//div[@class='mdk-body-detail']"):
     print(results.text)
```

```python
results = driver.find_elements_by_xpath("//div[@class='mdk-body-detail']")
len(results)
```

```python
Traveling = []
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
for product in driver.find_elements_by_xpath("//div[@class='mdk-body-detail']"):
    if product.text != '':
        images = []
        for img in product.find_elements_by_tag_name('img'):
            images.append({'alt': img.get_attribute('alt'), 'src': img.get_attribute('src')})
        Traveling.append({
            'brand_img': images[0]['src'],
        })
```
